# Remove columns in Dataframe 

```python
columns_to_remove = [
    "Unnamed: 0",
    "S.No.",
    "prior",
    "vice",
    
]

presidents= presidents.drop(columns=columns_to_remove)
presidents
```

#Remove rows forom the DataFrame and reset the index after dropping roews 

```python
import pandas as pd

presidents = presidents.drop(range(41))

presidents = presidents.reset_index(drop=True)
```

```python
presidents
```

#Replace NaN my new value 

```python
import pandas as pd


presidents.replace('--', pd.NA, inplace=True)


presidents.fillna(value='January 20,2017', inplace=True)

```
